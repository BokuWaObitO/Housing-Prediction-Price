## Custom Transformers (Weak)

1.Imports:<br>
*ndarray*: is imported from numpy, which is a type used to represent arrays in NumPy.<br>
*BaseEstimator & TransformerMixin*: are imported from sklearn.base. These are base classes provided by scikit-learn to create custom transformers that can be used in scikit-learn pipelines.<br>

2.Index Constants:<br>

*rooms_ix, bedrooms_ix, population_ix, households_ix* are index constants set to 3, 4, 5, and 6, respectively. These are used to access specific columns in the dataset. The indices correspond to the columns for rooms, bedrooms, population, and households in the dataset.<br>

3.CombinedAttributeAdder Class:
This class inherits from BaseEstimator and TransformerMixin, making it compatible with scikit-learn's pipeline and grid search functionalities.<br>

a.*__init__ Method*:
Takes an argument add_bedrooms_per_room (default is True){??Flexibility:
By providing a default value of True, the code allows the class to be instantiated without explicitly specifying this parameter.<br>

Users of the class can override this default by passing False when creating an instance if they don't want this feature.# Using default value (True)
obj1 = ClassName()

 Explicitly setting to False
obj2 = ClassName(add_bedrooms_per_room=False)
}<br>.

 This parameter determines whether the transformer will compute and add a new feature called bedrooms_per_room.<br>

Initializes an instance variable self.add_bedrooms_per_room with the given argument.,<br>

b.*fit Method*:
This method is required by scikit-learn's transformer interface. It takes the input data X and an optional target y.<br>
It returns self, which is a common practice to allow method chaining but does not perform any fitting in this case since the transformer is stateless.<br>

c.*transform Method*:

Takes the input data X and an optional target y.
Computes two new features:<br>

*rooms_per_household*: The ratio of the number of rooms to the number of households.<br>

*population_per_household*: The ratio of the population to the number of households.<br>
If add_bedrooms_per_room is True, it also computes:<br>

*bedrooms_per_room*: The ratio of the number of bedrooms to the number of rooms.<br>

*Uses np.c_[] to concatenate these new features to the original dataset X. np.c_[] is a NumPy function used to concatenate arrays along the second axis (columns).*<br>

Returns the augmented dataset with the new features.<br>

These calculations are done using NumPy array slicing, where X[:, index] accesses all rows for the specified column index



```python
from numpy import ndarray
from sklearn.base import BaseEstimator ,TransformerMixin
rooms_ix,bedrooms_ix,population_ix,households_ix=3,4,5,6
class CombinedAttributeAdder(BaseEstimator,TransformerMixin):
    def __init__(self,add_bedrooms_per_room=True):
        self.add_bedrooms_per_room=add_bedrooms_per_room
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        rooms_per_household=X[:,rooms_ix]/X[:,households_ix]
        population_per_household=X[:,population_ix]/X[:,households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room=X[:,bedrooms_ix]/X[:,rooms_ix]
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
        else:# mistake made was that its 'np.c_' not np.c
            return np.c_[X,rooms_per_household,population_per_household] 
attri_adder=CombinedAttributeAdder(add_bedrooms_per_room=False)
housing_extra_attrib=attri_adder.transform(housing.values)        
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline=Pipeline([
    ("imptr",SimpleImputer(strategy="median")),
    ("attrib_adder",CombinedAttributeAdder()),
    ("std_scal",StandardScaler()),
])# always remember to use " , " at the last list step , here after std_scaler last line 

housing_num_tr=num_pipeline.fit_transform(housing_num)

#Column Trasnfromers

from sklearn.compose import ColumnTransformer
#This line imports the ColumnTransformer class from the sklearn.compose module. ColumnTransformer is used to apply different preprocessing steps to different columns of a dataset.
num_attribs=list(housing_num)
cat_attribs=["ocean_proximity"]

full_pipeline=ColumnTransformer([
    ("num",num_pipeline,num_attribs),
    ("cat",OneHotEncoder(),cat_attribs),
    ])

housing_prep=full_pipeline.fit_transform(housing)
```

d.**scoring="neg_mean_squared_error"**:This specifies the scoring method to evaluate the predictions. The negative mean squared error is used because cross_val_score expects a score where higher is better. By using the negative MSE, you can later convert it to a positive RMSE.<br> 

e.**cv=10**: This specifies the number of cross-validation folds. The data will be split into 10 subsets, and the model will be trained and evaluated 10 times, each time using a different subset as the validation set and the remaining data as the training set.

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(tree_reg,housing_prep,housing_labels,scoring="neg_mean_squared_error",cv=10)

tree_rmse_scores=np.sqrt(-scores)
#-scores cause the scores are negative mean squared errors, negating them gives the positive mean squared errors.
# check to run this cell , u will see it took around 2 sec to compute the result ,that shows that cross-validation comes at the cost of training
#the model several times, so it is not always possible.
```

```python
def display_scores(scores):
    print(f"Scores:{scores}\n")
    print(f"Mean:{scores.mean()}\n")
    print(f"Std_dev:{scores.std()}\n")

display_scores(tree_rmse_scores) 
```
